## Step 3: Data Cleansing and Transformation

### Removal of Ouliers

```python
# Explore the distributuion of the variables

def diagnostic_plots(df, variable, title):
    # function takes a dataframe (df) and
    # the variable of interest as arguments

    # define figure size
    plt.figure(figsize=(16, 4))

    # histogram
    plt.subplot(1, 3, 1)
    sns.distplot(df[variable], bins=30)
    plt.title(f'{title} Histogram')

    # Q-Q plot
    plt.subplot(1, 3, 2)
    stats.probplot(df[variable], dist="norm", plot=plt)
    plt.ylabel('Quantiles')
    plt.title(f'{title} Q-Q Plot')

    # boxplot
    plt.subplot(1, 3, 3)
    sns.boxplot(y=df[variable])
    plt.title(f'{title} Boxplot')


    plt.show()
    
```

```python
# Display histogram, Q-Q plot and Box plot for numerical variables

for variable in ['bb', 'so', 'salary']:
    diagnostic_plots(df_baseball, variable, title = variable)
```

```python
# Trimming function 

def find_boundaries(df, variable, distance):

    # Let's calculate the boundaries outside which sit the outliers
    # for skewed distributions

    # distance passed as an argument, gives us the option to
    # estimate 1.5 times or 3 times the IQR to calculate
    # the boundaries.

    IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)

    lower_boundary = df[variable].quantile(0.25) - (IQR * distance)
    upper_boundary = df[variable].quantile(0.75) + (IQR * distance)

    return upper_boundary, lower_boundary
```

```python
# Finding the upper and lower limit for bb

bb_upper_limit, bb_lower_limit = find_boundaries(df_baseball, 'bb', 1.5)
bb_upper_limit, bb_lower_limit
```

```python
# let's flag the outliers in the data set for bb

outliers_bb = np.where(df_baseball['bb'] > bb_upper_limit, True,
                       np.where(df_baseball['bb'] < bb_lower_limit, True, False))
```

```python
# Outliers were trimmed for bb

df_baseball1 = df_baseball.loc[~outliers_bb ]

df_baseball.shape, df_baseball1.shape
```

```python
# Finding the upper and lower limit for so

so_upper_limit, so_lower_limit = find_boundaries(df_baseball1, 'so', 1.5)
so_upper_limit, so_lower_limit
```

### Introduction with Value Based Problem Statement

In professional baseball, the home run is one of the most significant events in the game. It not only provides a boost to a team's score but also can shift the momentum of a game. As such,  accurately predicting home runs can provide strategic advantages to teams, coaches and analysts. Given the wealth of available data based on historical performance and other relevant features, this presents an ideal opportunity to develop predictive models for the number of home runs. This model will help management make informed decisions about hiring the most talented and elite baseball players.

### Problem Formulation

The 'player', 'pitching', and 'salary' datasets were selected and explored: in 'player', we focused on the physical statistics ('height' and 'weight'); in 'pitching', we examined pitching statistics ('hr' for home run, 'bb' for base on balls, 'so' for strikeouts and 'throws' for laterality); and in 'salary', we analysed the 'salary' column to understand player compensation. The missing data analysis revealed that 'height' had 805 missing values, 'weight' had 871 missing values and there were no null values in 'bb', 'so', and 'salary'. The columns 'year', 'team_id' and 'league_id' were present in both the 'pitching' and 'salary' datasets. If the data in both columns are identical, the duplicated columns can be dropped before merging. If the data in both columns are unique, they should be renamed to prevent confusion. However, since these features were not utilised, this step was not performed in the assignment. To determine if the model can predict home runs based on physical statistics, pitching statistics and salary, home runs will be designated as the target variable while the remaining features will serve as the independent variables.

### Data Wrangling on multiple tables, Data Cleansing and Transformation

The 'player', 'pitching' and 'salary' datasets were merged with an inner join on 'primary_id' as the primary key. No null values were found on the columns of interest during the analysis after the merge. Five of these are numerical variables: 'height', 'weight', 'hr', 'bb', and 'so', while one is a categorical variable: 'throws'. 

We further investigated the linear relationship between the selected numerical variables. There is a strong positive relationship between 'bb', 'so' and 'salary' and the target variable. Interestingly, although 'height' and 'weight' were labelled as numerical variables, they appeared segmented in the plots because the data is recorded without decimal places. Consequently, they will be further processed and discretised before training the model.

Histograms, Q-Q plots, and box plots were created for 'bb', 'so' and 'salary' to analyse their distributions and detect any outliers. Outliers in the dataset were addressed through trimming and since all three columns' distributions were positively skewed, different transformation techniques were applied to make the data appear more normally distributed. Yeo-Johnson was applied for 'bb' and 'so' while Power-transformation was utilised to transform 'salary. During the train-test split, the data set except the target variable was allocated such that 70% of the data was used for the training set and 30% for the test set. The 'height' and 'weight' columns were discretised using equal frequency discretisation to create ranges and then encoded ordinally to establish a monotonic relationship. For the 'throws' column which is categorical, one-hot encoding was applied.

### Machine Learning Model

The final dataframe prepared for model training consists of 82230 rows and 7 columns. A naive baseline was build to establish a baseline to compare with more complex linear regression models. The Naive Baseline Model's test data metrics revealed MSE of 59.49  RMSE of 7.71, and an R² of -0.00. The initial Linear Regression Model's test data metrics revealed MSE of 24.19, an RMSE of 4.92, and an R² of 0.59. The data was further refined and better results were achieved when 'bb', 'so', and 'salary' were left untransformed after outliers were removed. With this approach, the Linear Regression Model's test data metrics showed an MSE of 21.08, an RMSE of 4.59, and an R² of 0.64. Over transformation might lead to some overfitting which has resulted in the slight underperformance.

### Summary and Further Improvements

Overall, the linear regression model demonstrates better performance in terms of both error metrics and variance compared to the naive baseline model as it consistently achieves lower MSE and RMSE values and demonstrates a moderate relationship with the home run counts, as indicated by its R² value. 

While the current model provides valuable insights into predicting home run counts, there is potential for further improvement. The Linear Regression Model’s RMSE of 4.92 indicates that the prediction error is relatively high compared to the mean value of home run in the dataset of about 9.27. The model's R² value of 0.64 indicates that it explains a good portion of the variance in home run performance but suggests there is still some variability that could be accounted for with additional or improved features. 

Incorporating additional relevant features can improve the model’s capacity to identify and understand the underlying patterns in the data. This can potentially lead to better predictions if the new features provide additional useful information about the target variable. On top of that, we can also use an outer join to prevent data loss which can help in maintaining the integrity and completeness of the dataset used for training and testing the model. Train-test split should also be done before data preprocessing to prevent data leakage.
